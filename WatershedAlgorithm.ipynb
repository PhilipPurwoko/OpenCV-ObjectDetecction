```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pennies = cv2.imread('Images/pennies.jpg')

# Need to blur the image
pennies = cv2.medianBlur(pennies,35)
pennies = cv2.cvtColor(pennies,cv2.COLOR_BGR2RGB)
# Gray copy
gray_pennies = cv2.cvtColor(pennies.copy(),cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(pennies)
plt.show()
plt.imshow(gray_pennies,cmap='gray')
```

```python
# Threshold image
_,threshold_pennies = cv2.threshold(gray_pennies,0,255,cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
```

```python
plt.imshow(threshold_pennies,cmap='gray')
```

```python
# Remove noise (just in case)
kernel = np.ones((3,3),np.uint8)
```

```python
clear_threshold_pennies = cv2.morphologyEx(threshold_pennies,cv2.MORPH_OPEN,kernel,iterations=2)
```

```python
plt.imshow(clear_threshold_pennies,cmap='gray')
```

```python
# Find sure background
sure_background = cv2.dilate(clear_threshold_pennies,kernel,iterations=3)
```

```python
plt.imshow(sure_background,'gray')
```

```python
# Distance transform
distance_transform = cv2.distanceTransform(clear_threshold_pennies,cv2.DIST_L2,5)
```

```python
plt.imshow(distance_transform,cmap='gray')
```

```python
# Find sure foreground via distance transform
_, sure_foreground = cv2.threshold(distance_transform,0.7*distance_transform.max(),255,0)
sure_foreground = np.uint8(sure_foreground)
```

```python
plt.imshow(sure_foreground,'gray')
```

```python
# Sure Background - Sure Foreground = Unknown
unknown = cv2.subtract(sure_background,sure_foreground)
```

```python
plt.imshow(unknown,'gray')
```

```python
# Create Marker
_,marker = cv2.connectedComponents(sure_foreground)
marker += 1
marker[unknown == 255] = 0
```

```python
# Solid Gray = Background
# Solid Black = Unknown
# Grayscale = Foreground
plt.imshow(marker,'gray')
```

```python
# Use Watershed Algorithm
watershed_marker = cv2.watershed(pennies,marker)
```

```python
plt.imshow(watershed_marker,'gray')
```

```python
# Detect Contour
# cv2.RETR_CCOMP = Mode
# cv2.CHAIN_APPROX_SIMPLE = Method
contours, hierarchy = cv2.findContours(watershed_marker,cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)
```

```python
for i in range(len(contours)):
    if hierarchy[0][i][3] == -1:
        cv2.drawContours(pennies,contours,i,(255,0,0),10)
```
