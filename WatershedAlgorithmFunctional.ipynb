```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def detectImage(image_location,blur_power=None,contour_thickness=10):
    fig,axes = plt.subplots(ncols=3,nrows=2,figsize=(20,10))
    plt.suptitle('Watershed Algorithm',fontsize=30)
    
    def axe_gen():
        for i in range(2):
            for j in range(3):
                yield axes[i,j]
    axeses = axe_gen()
    
    def plot(selected_axes,image,title,gray=True):
        if gray:
            selected_axes.imshow(image,cmap='gray')
        else:
            selected_axes.imshow(image)
        selected_axes.set_title(title)
        
    # Need to blur the image
    if blur_power == None:
        image = cv2.cvtColor(cv2.imread(image_location),cv2.COLOR_BGR2RGB)
    else:
        image = cv2.cvtColor(cv2.medianBlur(cv2.imread(image_location),blur_power),cv2.COLOR_BGR2RGB)
    plot(next(axeses),image,f'{image_location}',gray=False)
    
    # Gray copy
    gray_image = cv2.cvtColor(image.copy(),cv2.COLOR_RGB2GRAY)
    
    # Threshold image
    _,threshold_image = cv2.threshold(gray_image,0,255,cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
    
    # Remove noise (just in case)
    kernel = np.ones((3,3),np.uint8)
    clear_threshold_image = cv2.morphologyEx(threshold_image,cv2.MORPH_OPEN,kernel,iterations=2)
    plot(next(axeses),clear_threshold_image,'Threshold Image')
    
    # Find sure background
    sure_background = cv2.dilate(clear_threshold_image,kernel,iterations=3)
    
    # Distance transform
    distance_transform = cv2.distanceTransform(clear_threshold_image,cv2.DIST_L2,5)
    plot(next(axeses),distance_transform,'Distance Transform')
    
    # Find sure foreground via distance transform
    _, sure_foreground = cv2.threshold(distance_transform,0.7*distance_transform.max(),255,0)
    sure_foreground = np.uint8(sure_foreground)
    
    # Sure Background - Sure Foreground = Unknown
    unknown = cv2.subtract(sure_background,sure_foreground)
    
    # Create Marker
    _,marker = cv2.connectedComponents(sure_foreground)
    marker += 1
    marker[unknown == 255] = 0
    plot(next(axeses),marker,'Marker Image',gray=False)
    
    # Use Watershed Algorithm
    watershed_marker = cv2.watershed(image,marker)
    plot(next(axeses),watershed_marker,'Watershed Image',gray=False)
    
    # Detect Contour
    contours, hierarchy = cv2.findContours(watershed_marker,cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)
    for i in range(len(contours)):
        if hierarchy[0][i][3] == -1:
            cv2.drawContours(image,contours,i,(255,0,0),contour_thickness)
        
    plot(next(axeses),image,'Final Detection')
```

```python
detectImage('Images/pennies.jpg',blur_power=35,contour_thickness=30)
```

```python
detectImage('Images/internal_external.png',blur_power=None,contour_thickness=5)
```
