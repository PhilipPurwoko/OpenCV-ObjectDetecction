```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
source = cv2.imread('Images/internal_external.png',0) # 0 to use grayscale
source.shape
```

```python
plt.imshow(source,cmap='gray')
```

```python
def detectContour(source,external=True,only_final=False):
    contour,hierarchy = cv2.findContours(source,cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)
    desired_contour = np.zeros(source.shape)
    
    def drawContour(ext_int):
        cv2.drawContours(desired_contour,contour,i,255,-1)
        plt.imshow(desired_contour,cmap='gray')
        plt.title(f'Found {ext_int} contour in hirearchy number {i}')
        
        if only_final:
            pass
        else:
            plt.show()
    
    for i in range(len(contour)):
        if external:
            if hierarchy[0][i][-1] == -1:
                drawContour('external')
        else:
            if hierarchy[0][i][-1] != -1:
                drawContour('internal')
```

```python
detectContour(source,external=False,only_final=True)
```
