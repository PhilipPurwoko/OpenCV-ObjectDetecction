```python
import cv2
import matplotlib.pyplot as plt
```

```python
full = cv2.imread('Images/full.jpg')
full = cv2.cvtColor(full,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(full)
```

```python
face = cv2.imread('Images/face.jpg')
face = cv2.cvtColor(face,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(face)
```

```python
methods = ['cv2.TM_CCOEFF','cv2.TM_CCOEFF_NORMED','cv2.TM_CCORR','cv2.TM_CCORR_NORMED','cv2.TM_SQDIFF','cv2.TM_SQDIFF_NORMED']
```

```python
def detectViaTemplate(full_image,wanted_detection,list_of_method):
    for method in list_of_method:
        copy = full_image.copy()
        result = cv2.matchTemplate(copy,wanted_detection,eval(method)) # Returns heatmap
        
        # Find the coords (x,y)
        min_val,max_val,min_loc,max_loc = cv2.minMaxLoc(result)
        if method == 'cv2.TM_SQDIFF' or method == 'cv2.TM_SQDIFF_NORMED':
            top_left = min_loc
        else:
            top_left = max_loc
        
        height,width,channel = face.shape
        bottom_right = (top_left[0]+width,top_left[1]+height)
        
        # Draw rectangle
        cv2.rectangle(copy,top_left,bottom_right,(255,0,0),30)
        
        # Plot them
        fig, axes = plt.subplots(nrows=1,ncols=2)
        axes[0].imshow(copy,)
        axes[0].set_title('Detection')
        axes[1].imshow(result)
        axes[1].set_title('Heatmap')
        
        fig.suptitle(f'Method use : {method}')
        plt.show()
```

```python
detectViaTemplate(full,face,methods)
```
