```python
import cv2
import cvlib
import matplotlib.pyplot as plt
```

```python
draw_box = cvlib.object_detection.draw_bbox
```

```python
def detect(image):
    bbox, label, confidence = cvlib.detect_common_objects(image)
    return draw_box(image,bbox,label,confidence)
```

```python
videoCapture = cv2.VideoCapture(0)
while True:
    response, frame = videoCapture.read()

    cv2.imshow('Camera',detect(frame))
    
    if cv2.waitKey(1) == 27:
        break
videoCapture.release()
cv2.destroyAllWindows()
```

```python
image = cv2.imread('Images/Apple-Books.jpeg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
figure,axes = plt.subplots(ncols=2,nrows=1,figsize=(20,10))
axes[0].imshow(image)
axes[1].imshow(detect(image))
```
